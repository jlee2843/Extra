Hmm, does this look right? No, we can't interpret this bar graph. We need to calculate the average math score of each parent education level separately. How can we do this? 

Solution: Lets use the pipe operator %>%, and the functions group_by and summarize. 
%>% is used to apply a sequence of functions 

```r
# calculate summary statisitcs 

df_math <- df %>% group_by(parent_edu_level) %>%
    summarize(av_math_score = mean(math_score))
```

```r
# another example calculating median via summarize

df_math_median <- df %>% group_by(parent_edu_level) %>%
    summarize(med_math_score = median(math_score))

df_math_median
```

Since there is a strong linear relationship here, can we use someones reading score to predict writting score? 

Lets fit a linear model! 
R has a built in function lm()

```r
# simple linear regression model 

reg <- lm(writting_score ~ reading_score, data=df)
```

Simple linear regression minimizes the sum of squares of prediction error to get the least square regression line

 minimize $\Sigma(y_{i} - b_{0} - b_{1}x)^2$

$\hat{y} = b_{0} + b_{1}x_{i}$

$b_{0}$ is the intercept $b_{1}$ is the coefficient learned by the model 

we can use the summary(reg) to look at our model

```r
summary(reg)
```

How can we tell if our model is good? 


- look at R^2 value 

$R^2$ represents the percentage of the variance of the response variable that can be explained by the predictor variables of our model. It is a measure of how good our regression line fits the data. 

perfect $R^2$ value is 1

```r
summary(reg)$r.squared 
```

Wow! Since this is very close to 1, it means our model is a good fit

What if we want to add in more predictor varibles? Can R handle categorical variables such as test_prep_course ? 

```r
reg2 <- lm(writting_score ~ reading_score + test_prep_course, data=df)
```

```r
summary(reg2)
```

For binary variables, R uses one as a baseline, which is why we only see test_prep_coursenone above. 

How do we know if a coefficient is important? 
1. does adding it increase adjusted R^2 value? 
2. is it significant? 

We know whether a coeefficient is significant by looking at it's p-value in the summary output. 

Test: $H_{0}$: coefficient = 0 vs $H_{a}$: coefficient $\neq$ 0

Test statistic: $\frac{B_{i}-\beta_{i}}{se(B_{i})}$ ~ $t_{n-k}$


Thankfully R does this for us! Look at the P(>|t|) in the summary ouput for significance! We see that all the coefficients are significant at alpha=0.001.

```r
summary(reg2)
```
